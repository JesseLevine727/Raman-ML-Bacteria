```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = 'PAO_Normalized_and_Outlier_Corrected_Data.csv'
df = pd.read_csv(file)
```

```python
x_axis = df.iloc[0, :-1].index
data_rows = df.iloc[1:, :-1]

x_axis
```

```python
plt.figure(figsize= (12,8))

for index, row in data_rows.iterrows():
    plt.plot(x_axis, row.values)

plt.title('Normalized and Outlier Corrected C. Striatum Spectra')
plt.xlabel('Wavenumber cm^{-1}')
plt.ylabel('Min-Max au')
#plt.grid(True)    
plt.show()
```

```python
median_values = data_rows.median()

plt.figure(figsize=(12, 8))
plt.plot(x_axis, median_values.values)

plt.title('Median Plot')
plt.xlabel('Wavenumber (cm^{-1})')
plt.ylabel('Median Values')

plt.show()
```
